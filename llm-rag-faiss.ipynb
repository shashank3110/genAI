```python
import faiss
from sentence_transformers import SentenceTransformer
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
```

```python
df_sub = df.head(n=500).to_pandas()
df_raw_sub = df_raw.head(n=500).to_pandas()
```

### Load Embedding model and create embeddings for text data

```python
# Load the SentenceTransformer model for embedding creation
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def create_embeddings(text_list, model):
    embeddings = model.encode(text_list, convert_to_numpy=True)
    return embeddings
```

```python
# recipe_name embeddings

name_embeddings = create_embeddings(df_raw_sub['name'].to_list(), embedding_model)
```

```python
# recipe_steps emebddings

steps_embeddings = create_embeddings(df_raw_sub['steps'].to_list(), embedding_model)
```

```python
# recipe_ingredients emebddings

ingredients_embeddings = create_embeddings(df_raw_sub['ingredients'].to_list(), embedding_model)
```

### Create Indices and embeddings (using FAISS-Facebook AI Semantic search)

```python
# Normalize embeddings for FAISS and Create index


faiss.normalize_L2(name_embeddings)
faiss.normalize_L2(steps_embeddings)
faiss.normalize_L2(ingredients_embeddings)

name_index = faiss.IndexFlatL2(name_embeddings.shape[1])
steps_index = faiss.IndexFlatL2(steps_embeddings.shape[1])
ingredients_index = faiss.IndexFlatL2(ingredients_embeddings.shape[1])
```

```python
# Add embeddings to FAISS Indices

name_index.add(name_embeddings)
steps_index.add(steps_embeddings)
ingredients_index.add(ingredients_embeddings)
```

### Load LLM, Create RAG Query and Query the LLM

```python
# Load LLM


llm = pipeline("text-generation", model="distilgpt2", device=-1)




```

```python
help(ingredients_index.search)
```

```python


def get_rag_query(query, top_k=5):
    query_embedding = embedding_model.encode([query])
    faiss.normalize_L2(query_embedding)
    
    # using the query find top relevant indegresdients and recipes
    name_dist, top_names = name_index.search(query_embedding, top_k)
    steps_dist, top_steps = steps_index.search(query_embedding, top_k)
    ing_dist, top_ingredients = ingredients_index.search(query_embedding, top_k)
    print(top_steps[0])
    ingredients_metadata = [f"Ingredient {df_raw_sub['ingredients'][i]}" for i in top_ingredients[0]]
    recipe_name_metadata = [f"Recipe name {df_raw_sub['name'][i]}" for i in top_names[0]]
    recipe_steps_metadata = [f"Recipe steps {df_raw_sub['steps'][i]}" for i in top_steps[0]]
    
    # Augment query
    augmented_query = (
        f"User Query: {query}\n\n"
        f"Top Ingredients: {', '.join(ingredients_metadata)}\n"
        f"Top Recipe Name: {', '.join(recipe_name_metadata)}\n"
        # f"Top Recipe Steps: {', '.join(recipe_steps_metadata)}\n\n" ## uncoment this if you want to give dataset recipes in llm input
        f"Method:"
    )


    return augmented_query
```

```python
query = "What can I cook with tomatoes and chicken?"
rag_query = get_rag_query(query, top_k=3)
```

```python
rag_query
```

```python
# Querry LLM with RAG

response = llm(rag_query, max_length=1024, num_return_sequences=1) # increasing max_length might give error depends on which llm is used
response[0]['generated_text']
```
