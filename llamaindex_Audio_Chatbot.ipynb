```python
import os
os.environ["OPENAI_API_KEY"] = "openai_api_key"
```

```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader, TreeIndex
from llama_index.tts import ElevenLabsTTS
from IPython.display import Markdown, display, Audio
```

```python
documents = SimpleDirectoryReader("sample_data").load_data()
index = TreeIndex.from_documents(documents)
```

```python
# querying the llm
query_engine = index.as_query_engine()
response = query_engine.query("When did Shashank graduate?")
```

```python
# text to speech - using Elevenlabs TTS
os.environ["ElevenLabs_key"] = "eleven_labs_api_key"
elevenlabs_key = os.environ["ElevenLabs_key"]
tts = ElevenLabsTTS(api_key=elevenlabs_key)
```

```python
str(response)
```

```python
# tts reponse
audio_response = tts.generate_audio(str(response))
```

```python
# display Audio response as cell output
Audio(audio_response)
```
